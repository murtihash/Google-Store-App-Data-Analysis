```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
rdf= pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
rdf.info()
```

```python
df= pd.read_csv('googleplaystore.csv')
```

```python
# Need to merge on 'APP' in order to be able to compare types of reviews for each category
```

```python
mdf= pd.merge(rdf, df, on="App",how="inner")
```

```python
mdf =mdf.drop_duplicates()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(y=mdf['Category'],hue=mdf['Sentiment'],orient='h')
plt.xticks(fontsize=20 , rotation=90)
plt.yticks(fontsize=20)
plt.legend(prop={'size': 30})
plt.title('Types of Reviews for each Category',fontdict={'fontsize':40})

```

```python
# Game Category appears to have the most Positive and Negative Reviews. Shows volatility of Game Category.
# Health and Fitness appears to have least Negative reviews in relation to its positive reviews.
```

```python
plt.figure(figsize=(10,8))
sns.stripplot(x=mdf['Type'],y=mdf['Sentiment_Polarity'],jitter=True)
plt.title('Free Vs Paid Sentiment Polarity',fontdict={'fontsize':40})
sns.set_style("whitegrid")
plt.rcParams["patch.force_edgecolor"]=True
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
# Clearly Paid Apps are less likely to recieve bad reviews
# Free apps have a more equal distribution of Setiment Polarity
```
